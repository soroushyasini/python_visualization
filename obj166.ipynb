## ploting pizo and subs data with linear interpolation

```python


# First plot
x1 = pizo['Tarikh_M']
y1 = pizo['Taraz_ab']

# Second plot
x2 = sub_data['date']
y2 = sub_data['obj166']

# Set a custom Seaborn style
sns.set_style("dark",{'grid.linestyle': ''})
sns.set_palette("husl")  # Custom color palette

# Create a figure and the first subplot
fig, ax1 = plt.subplots(figsize=(14, 4))

# Use Seaborn to plot the first dataset with a more attractive style
sns.lineplot(x=x1, y=y1, marker='o', markersize=10, label='Plot 1', ax=ax1, color='royalblue', linestyle='--', linewidth=2)

ax1.set_xlabel('Date')
ax1.set_ylabel('Water Level (Meter)')


# Create the second subplot sharing the same X-axis
ax2 = ax1.twinx()

# Use Seaborn to plot the second dataset with a different color and style
sns.lineplot(x=x2, y=y2, marker='s', markersize=8, label='Plot 2', ax=ax2, color='indianred', linestyle='--', linewidth=2)

ax2.set_ylabel('Subsidance (mm)')
ax2.get_yaxis().set_label_coords(1.12, 0.5)  # Adjust the position of the second y-axis label


# Add a title and adjust the layout
plt.title('Combined Plot with Dual Y-Axes', fontsize=16)
plt.tight_layout()
# Combine both legends into a single legend
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
combined_handles = handles1 + handles2
combined_labels = labels1 + labels2

# Display the combined legend
plt.legend(combined_handles, combined_labels, loc='upper right')
# Show the combined plot
plt.show()
```

## ploting pizo and subs data with trendline

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# First plot
x1 = pizo['Tarikh_M']
y1 = pizo['Taraz_ab']

# Second plot
x2 = sub_data['date']
y2 = sub_data['obj166']

# Set the Seaborn style and remove grid lines
sns.set_style("darkgrid", {'grid.linestyle': ''})

# Convert the dates to numerical values
x1_numeric = (x1 - x1.min()).dt.days
x2_numeric = (x2 - x2.min()).dt.days

# Create a figure and the first subplot
fig, ax1 = plt.subplots(figsize=(14, 4))

# Use Seaborn to plot the first dataset
sns.scatterplot(x=x1, y=y1, marker='o', s=50, color='royalblue', ax=ax1)

ax1.set_xlabel('Date')
ax1.set_ylabel('Water Level (Meter)')

# Calculate the trendline for the first dataset
z1 = np.polyfit(x1_numeric, y1, 1)
p1 = np.poly1d(z1)
plt.plot(x1, p1(x1_numeric), linestyle='--', color='royalblue', label='Trendline (Water Level)', linewidth=4)

# Create the second subplot sharing the same X-axis
ax2 = ax1.twinx()

# Use Seaborn to plot the second dataset
sns.scatterplot(x=x2, y=y2, marker='s', s=25, color='indianred', label='Subsidence', ax=ax2)

ax2.set_ylabel('Point Subsidence (Millimeter)')
ax2.get_yaxis().set_label_coords(1.12, 0.5)  # Adjust the position of the second y-axis label

# Calculate the trendline for the second dataset
z2 = np.polyfit(x2_numeric, y2, 1)
p2 = np.poly1d(z2)
plt.plot(x2, p2(x2_numeric), linestyle='--', color='indianred', label='Trendline (Subsidence)', linewidth=4)

# Combine both legends into a single legend, excluding "Water Level" label
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
combined_handles = handles1 + handles2
combined_labels = labels1 + labels2
combined_handles = [h for h, label in zip(combined_handles, combined_labels) if label != 'Water Level']
combined_labels = [label for label in combined_labels if label != 'Water Level']

# Display the combined legend
plt.legend(combined_handles, combined_labels, loc='best')

# Show the combined plot
plt.title('Pizometer and Subsidence')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# First plot
x1 = pizo['Tarikh_M']
y1 = pizo['Taraz_ab']

# Second plot
x2 = sub_data['date']
y2 = sub_data['obj166']

# Set the Seaborn style and remove grid lines
sns.set_style("darkgrid", {'grid.linestyle': ''})

# Convert the dates to numerical values
x1_numeric = (x1 - x1.min()).dt.days
x2_numeric = (x2 - x2.min()).dt.days

# Create a figure and the first subplot
fig, ax1 = plt.subplots(figsize=(14, 4))

# Use Seaborn to plot the first dataset
sns.scatterplot(x=x1, y=y1, marker='o', s=50, color='royalblue', ax=ax1)

ax1.set_xlabel('Date')
ax1.set_ylabel('Water Level (Meter)')

# Calculate the trendline for the first dataset (2015-2017)
mask1 = (x1 >= '2015-01-01') & (x1 <= '2018-12-31')
x1_range1 = x1[mask1]
y1_range1 = y1[mask1]
z1_range1 = np.polyfit((x1_range1 - x1_range1.min()).dt.days, y1_range1, 1)
p1_range1 = np.poly1d(z1_range1)
plt.plot(x1_range1, p1_range1((x1_range1 - x1_range1.min()).dt.days), linestyle='--', color='green', label='Trendline (2015-2017)', linewidth=4)

# Calculate the trendline for the first dataset (2019-2022)
mask2 = (x1 >= '2019-08-01') & (x1 <= '2022-12-31')
x1_range2 = x1[mask2]
y1_range2 = y1[mask2]
z1_range2 = np.polyfit((x1_range2 - x1_range2.min()).dt.days, y1_range2, 1)
p1_range2 = np.poly1d(z1_range2)
plt.plot(x1_range2, p1_range2((x1_range2 - x1_range2.min()).dt.days), linestyle='--', color='orange', label='Trendline (2019-2022)', linewidth=4)

# Create the second subplot sharing the same X-axis
ax2 = ax1.twinx()

# Use Seaborn to plot the second dataset
sns.scatterplot(x=x2, y=y2, marker='s', s=25, color='indianred', label='Subsidence', ax=ax2)

ax2.set_ylabel('Point Subsidence (Millimeter)')
ax2.get_yaxis().set_label_coords(1.12, 0.5)  # Adjust the position of the second y-axis label

# Calculate the trendline for the second dataset
z2 = np.polyfit(x2_numeric, y2, 1)
p2 = np.poly1d(z2)
plt.plot(x2, (p2(x2_numeric)-5), linestyle='--', color='indianred', label='Trendline (Subsidence)', linewidth=4)

# Convert the specified date to the same date format as x-axis data
vertical_line_start_date = pd.to_datetime('2019-01-10')
vertical_line_end_date = pd.to_datetime('2019-05-10')
# Add a vertical line at the specified date (2020-01-01)
plt.axvline(x=vertical_line_start_date, color='gray', linestyle='--', label='97 flood')
plt.axvline(x=vertical_line_end_date, color='gray', linestyle='--')

# Combine both legends into a single legend, excluding duplicate labels
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
combined_handles = handles1 + handles2
combined_labels = labels1 + labels2
unique_labels = []
unique_handles = []
for label, handle in zip(combined_labels, combined_handles):
    if label not in unique_labels:
        unique_labels.append(label)
        unique_handles.append(handle)

# Display the combined legend
plt.legend(unique_handles, unique_labels, loc='best')

# Show the combined plot
plt.title('Pizometer and Subsidence')
plt.show()
```
